# Moonshot Backtest

```python
default_params = {
    "MOMENTUM_WINDOW": 100, "VOLATILITY_WINDOW": 40, "RANKING_PERIOD_GAP": 10, 
    "TOP_N_PCT": None, "TOP_N_COUNT": 4, 
    "REBALANCE_INTERVAL": "M", "PRICE_LOWER_LIMIT": 500,
    "LONG_ONLY": False, "SHORT_ONLY": False, "ESCAPE_WEEKLY_CHANGE_LIMIT": 0.0,
    "DB": "usstock-1d", "UNIVERSES": "usstock"}
```

```python
from quantrocket.moonshot import backtest
backtest("momentum-score", start_date="2019-01-01", end_date="2023-01-12", 
         params = {**default_params},
         filepath_or_buffer="momentum_score_moonshot_results.csv")
from moonchart import Tearsheet
Tearsheet.from_moonshot_csv("momentum_score_moonshot_results.csv")
```

```python
Tearsheet.from_moonshot_csv("momentum_score_moonshot_results.csv", trim_outliers=20)
```

```python
from quantrocket.moonshot import backtest
backtest("momentum-score", start_date="2019-01-01", end_date="2023-01-12", 
         params = {**default_params, "SHORT_ONLY": True, "EWM_COM": 4},
         filepath_or_buffer="momentum_score_moonshot_results.csv")
from moonchart import Tearsheet
Tearsheet.from_moonshot_csv("momentum_score_moonshot_results.csv", trim_outliers=20)
```

```python
from quantrocket.moonshot import scan_parameters
scan_parameters("momentum-score", 
                param1="EWM_COM",
                vals1=[None, 2, 4, 8],
                params = default_params,
                start_date="2019-01-01",
                end_date="2022-01-12",
                filepath_or_buffer="momentum_score_moonshot_EWM_COM.csv")
from moonchart import ParamscanTearsheet
ParamscanTearsheet.from_csv("momentum_score_moonshot_EWM_COM.csv")
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
_regr = linear_model.LinearRegression()

def _get_r2_score(y, days):
    '''
    regression score, the higher the more linear the momentum.
    '''
    global _regr
    l = len(y)
    if days // 2 > l:
        return 0

    y = y[-days:]
    x = np.array([i for i in range(len(y))]).reshape(-1, 1)
    y = np.array(y).reshape(-1, 1)
    _regr.fit(x, y)
    y_pred = _regr.predict(x)
    return r2_score(y, y_pred)

def _get_return(y, days):
    l = len(y)
    if l == 0:
        print('length is zero')
        return 0
    head = max(0, l - days)
    return (y[-1] - y[head]) / y[head]

def _get_momentum_score(y, days):
    ret =  _get_return(y, days) * _get_r2_score(y, days)
    if type(ret) is np.ndarray:
        return ret[0]
    return ret
```

```python
MOMENTUM_SCORE_DAYS = [20, 40, 60]

def get_momentum_df(df, intervals):
    # m score
    m_scores_per_symbol = defaultdict(list)
    sids = df.columns.values
    for sid in sids:
        for interval in intervals:
            try:
                m_scores_per_symbol[sid].append(_get_momentum_score(df[sid].dropna().values, interval))
            except KeyError as e:
                print(e)

    recent_date = df.index.get_level_values(1)[-1]
    df_recent = df.xs(recent_date, level=1)
    df_mscores = pd.DataFrame.from_dict(m_scores_per_symbol, orient='index', columns=['m_score{d}'.format(d=m_day) for m_day in intervals])
    #df_mscores = df_mscores.set_index(df_mscores.index.rename('sid')).join(df_recent).dropna()
    df_mscores = df_mscores.set_index(df_mscores.index.rename('sid'))
    df_mscores['Close'] = df_recent.T
    return df_mscores
```

## momentum

```python
def get_return(df, interval):
    return (df - df.shift(interval)) / df.shift(interval)

def get_momentum(df, interval_return, interval_vol):
    ret = get_return(df, interval_return)
    daily_return = df/df.shift(1) - 1
    vol = daily_return.rolling(interval_vol).std()
    momentum = ret / (0.05 + vol)
    return momentum
```
